```python
#@title Setup (run once)
import sys
import os

if "google.colab" in sys.modules:
    from google.colab import auth
    auth.authenticate_user()
    if not os.path.exists("lla-data"):
        !git clone -q https://github.com/aidoanto/lla-data.git
    repo = os.path.abspath("lla-data")
    if repo not in sys.path:
        sys.path.insert(0, repo)
    !pip install -q db-dtypes google-cloud-bigquery kaleido plotly
else:
    for p in ("..", "../.."):
        ap = os.path.abspath(p)
        if ap not in sys.path:
            sys.path.insert(0, ap)

import pandas as pd
import plotly.express as px
import plotly.io as pio

import lifeline_theme
from lla_data import config
from lla_data.bq import default_query_window, get_client, load_sql_template, run_query

# Ensure notebook render context has brand fonts loaded.
lifeline_theme.inject_fonts()
# Make lifeline the default so charts inherit theme even if template is omitted.
pio.templates.default = "lifeline"

client = get_client()
```

```python
#@title Parameters
DAYS_BACK = 28 #@param {type:"integer"}

window = default_query_window(days_back=DAYS_BACK)

print(f"Project: {config.PROJECT_ID}")
print(f"Datasets: GA4={config.GA4_DATASET}, GSC={config.SEARCHCONSOLE_DATASET}")
print(f"Requested date range: {window.start_date} to {window.end_date}")
```

```python
from google.cloud import bigquery

from lla_data.bq import load_sql_template

freshness_query = load_sql_template(
    "sql/notebooks/seo_search_freshness.sql",
    project_id=config.PROJECT_ID,
    searchconsole_dataset=config.SEARCHCONSOLE_DATASET,
    ga4_dataset=config.GA4_DATASET,
)

freshness = run_query(client, freshness_query).iloc[0]
source_max_dates = [freshness["gsc_max_date"], freshness["ga4_max_date"], freshness["seo_max_date"]]
source_max_dates = [d for d in source_max_dates if pd.notna(d)]

if source_max_dates:
    available_end = min(source_max_dates)
else:
    available_end = window.end_date

analysis_end = min(window.end_date, available_end)

print(f"Data available through: {analysis_end}")
if analysis_end < window.end_date:
    print("Note: requested range extends past currently loaded data. This is usually pipeline lag.")

query = load_sql_template(
    "sql/notebooks/seo_search_share_daily.sql",
    project_id=config.PROJECT_ID,
    searchconsole_dataset=config.SEARCHCONSOLE_DATASET,
)

params = [
    bigquery.ScalarQueryParameter("start_date", "DATE", window.start_date),
    bigquery.ScalarQueryParameter("end_date", "DATE", analysis_end),
]

df = run_query(client, query, params=params)

if not df.empty:
    date_spine = pd.DataFrame({
        "report_date": pd.date_range(window.start_date, analysis_end, freq="D").date
    })
    df = date_spine.merge(df, on="report_date", how="left")

suspect_zero_gsc = df[
    (df["organic_sessions"].fillna(0) > 0)
    & (df["gsc_impressions"].fillna(0) == 0)
    & (df["gsc_clicks"].fillna(0) == 0)
]
if not suspect_zero_gsc.empty:
    dates = ", ".join(suspect_zero_gsc["report_date"].astype(str).tolist())
    print(f"Potential data issue: organic sessions exist but GSC clicks/impressions are zero on: {dates}")

df.tail()
```

```python
# Re-inject fonts for reliability if outputs were cleared/reloaded.
lifeline_theme.inject_fonts()

df_viz = df.copy()
df_viz["search_sessions"] = df_viz["organic_sessions"].fillna(0)
df_viz["other_sessions"] = (
    df_viz["total_sessions"].fillna(0) - df_viz["search_sessions"]
).clip(lower=0)
df_viz["search_share"] = df_viz["search_sessions"].div(df_viz["total_sessions"]).fillna(0)
df_viz["other_share"] = (1 - df_viz["search_share"]).clip(lower=0, upper=1)

chart_start = pd.to_datetime(df_viz["report_date"]).min()
chart_end = pd.to_datetime(df_viz["report_date"]).max()
if pd.isna(chart_start) or pd.isna(chart_end):
    analysis_window_label = f"{window.start_date} to {analysis_end}"
else:
    analysis_window_label = f"{chart_start:%Y-%m-%d} to {chart_end:%Y-%m-%d}"

fig_mix = px.area(
    df_viz,
    x="report_date",
    y=["search_sessions", "other_sessions"],
    template="lifeline",
    title=f"Sessions by Audience (Search vs Other) with Total ({analysis_window_label})",
    labels={"value": "Sessions", "report_date": "Date", "variable": "Audience"},
)
fig_mix.for_each_trace(
    lambda trace: trace.update(
        name={
            "search_sessions": "Search sessions",
            "other_sessions": "Other sessions",
        }.get(trace.name, trace.name)
    )
)
fig_mix.add_scatter(
    x=df_viz["report_date"],
    y=df_viz["total_sessions"],
    mode="lines",
    name="Total sessions",
    line={"color": "#221FBB", "width": 2},
)
lifeline_theme.add_lifeline_logo(fig_mix)
fig_mix.show()

fig_share = px.area(
    df_viz,
    x="report_date",
    y=["search_share", "other_share"],
    template="lifeline",
    title=f"Session Mix Over Time (%) ({analysis_window_label})",
    labels={"value": "Share of sessions", "report_date": "Date", "variable": "Audience"},
)
fig_share.for_each_trace(
    lambda trace: trace.update(
        name={
            "search_share": "Search share",
            "other_share": "Other share",
        }.get(trace.name, trace.name)
    )
)
fig_share.update_yaxes(tickformat=".0%")
lifeline_theme.add_lifeline_logo(fig_share)
fig_share.show()
```

## GA4 Source/Medium Breakdown (Excluding Organic Search)

This section decomposes the current `other_sessions` bucket (defined earlier as `total_sessions - organic_sessions`) into clearer source/medium channel groups, then lists the top contributors.

Notes:
- This uses GA4 attribution data.
- Rows with `medium = organic` are excluded, so this is specifically a view of sessions excluding Organic Search.
- Reconciliation against `other_sessions` is included as a quick sanity check.

```python
from google.cloud import bigquery

TOP_OTHER_SOURCES = 12

traffic_table_check = load_sql_template(
    "sql/notebooks/ga4_curated_daily_traffic_exists.sql",
    project_id=config.PROJECT_ID,
    ga4_dataset=config.GA4_DATASET,
)

has_curated_daily_traffic = run_query(client, traffic_table_check).iloc[0]["table_count"] > 0
date_params = [
    bigquery.ScalarQueryParameter("start_date", "DATE", window.start_date),
    bigquery.ScalarQueryParameter("end_date", "DATE", analysis_end),
]

if has_curated_daily_traffic:
    other_query = load_sql_template(
        "sql/notebooks/ga4_curated_daily_traffic_excl_organic.sql",
        project_id=config.PROJECT_ID,
        ga4_dataset=config.GA4_DATASET,
    )
else:
    print("`curated_daily_traffic` not found. Falling back to GA4 `events_*` for this breakdown.")
    other_query = load_sql_template(
        "sql/notebooks/ga4_events_daily_traffic_excl_organic.sql",
        project_id=config.PROJECT_ID,
        ga4_dataset=config.GA4_DATASET,
    )

df_other_detail = run_query(client, other_query, params=date_params)

if df_other_detail.empty:
    print('No source/medium rows found for this date range after excluding Organic Search.')
else:
    # Re-inject fonts for reliability if outputs were cleared/reloaded.
    lifeline_theme.inject_fonts()

    # Daily bucket totals for stacked area chart.
    daily_bucket = (
        df_other_detail
        .groupby(['report_date', 'channel_bucket'], as_index=False)['sessions']
        .sum()
    )

    bucket_order = [
        'Direct (no identifiable referrer/campaign)',
        'Paid Media (CPC/PPC/Display/Paid Social)',
        'Organic Social',
        'Referral (links from other websites)',
        'Email',
        'Other / Unclassified',
    ]
    df_other_pivot = (
        daily_bucket
        .pivot(index='report_date', columns='channel_bucket', values='sessions')
        .fillna(0)
        .reset_index()
    )

    present_buckets = [b for b in bucket_order if b in df_other_pivot.columns]

    fig_other_mix = px.area(
        df_other_pivot,
        x='report_date',
        y=present_buckets,
        template='lifeline',
        title=f"Sessions Excluding Organic Search by Channel Bucket ({window.start_date} to {analysis_end})",
        labels={'value': 'Sessions', 'report_date': 'Date', 'variable': 'Channel bucket'},
    )
    lifeline_theme.add_lifeline_logo(fig_other_mix)
    fig_other_mix.show()

    # Top source/medium contributors across the same window.
    top_other = (
        df_other_detail
        .groupby(['source', 'medium'], as_index=False)['sessions']
        .sum()
        .sort_values('sessions', ascending=False)
    )
    total_excl_organic_search = top_other['sessions'].sum()
    top_other['share_excl_organic_search'] = top_other['sessions'].div(total_excl_organic_search).fillna(0)
    top_other['source_medium'] = top_other['source'] + ' / ' + top_other['medium']

    print(f'Top {TOP_OTHER_SOURCES} source/medium pairs (excluding Organic Search):')
    display(
        top_other[['source_medium', 'sessions', 'share_excl_organic_search']]
        .head(TOP_OTHER_SOURCES)
        .assign(share_excl_organic_search=lambda d: d['share_excl_organic_search'].map(lambda x: f"{x:.1%}"))
    )
```
