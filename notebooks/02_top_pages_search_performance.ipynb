```python
#@title Setup (run once)
import sys
import os

if "google.colab" in sys.modules:
    from google.colab import auth
    auth.authenticate_user()
    if not os.path.exists("lla-data"):
        !git clone -q https://github.com/aidoanto/lla-data.git
    repo = os.path.abspath("lla-data")
    if repo not in sys.path:
        sys.path.insert(0, repo)
    !pip install -U -q "plotly>=6.1.1" "kaleido>=1.2.0"
else:
    for p in ("..", "../.."):
        ap = os.path.abspath(p)
        if ap not in sys.path:
            sys.path.insert(0, ap)

import plotly.express as px

import lifeline_theme
from lla_data import config
from lla_data.bq import build_date_params, default_query_window, get_client, load_sql_template, run_query

lifeline_theme.inject_fonts()

client = get_client()
window = default_query_window(config.DEFAULT_DAYS_BACK)
```

```python
#@title Parameters
TOP_N = 25 #@param {type:"integer"}
```

```python
from google.cloud import bigquery

query = load_sql_template(
    "sql/notebooks/seo_top_pages_search_performance.sql",
    project_id=config.PROJECT_ID,
    searchconsole_dataset=config.SEARCHCONSOLE_DATASET,
)

params = build_date_params(window) + [
    bigquery.ScalarQueryParameter("top_n", "INT64", TOP_N),
]

df_pages = run_query(client, query, params=params)
df_pages.head()
```

```python
MAX_PATH_LABEL_LEN = 75
df_pages_plot = df_pages.copy()
df_pages_plot["page_path_label"] = df_pages_plot["page_path"].astype(str).apply(
    lambda p: p if len(p) <= MAX_PATH_LABEL_LEN else p[: MAX_PATH_LABEL_LEN - 1] + "\u2026"
)

analysis_window_label = f"{window.start_date} to {window.end_date}"

fig_clicks = px.bar(
    df_pages_plot.sort_values("clicks", ascending=True),
    x="clicks",
    y="page_path_label",
    orientation="h",
    hover_name="page_path",
    hover_data={"page_path_label": False},
    template="lifeline",
    title=f"Top {TOP_N} Pages by Search Clicks ({analysis_window_label})",
)
fig_clicks.update_layout(height=700, margin={"l": 420, "r": 40, "t": 80, "b": 40})
fig_clicks.update_yaxes(automargin=True)
lifeline_theme.add_lifeline_logo(fig_clicks)
fig_clicks.show()

fig_ctr = px.scatter(
    df_pages,
    x="avg_position",
    y="ctr",
    size="impressions",
    hover_name="page_path",
    template="lifeline",
    title=f"CTR vs Average Position (bubble size = impressions) ({analysis_window_label})",
)
fig_ctr.update_yaxes(tickformat=".0%")
fig_ctr.add_annotation(
    xref="paper",
    yref="paper",
    x=0,
    y=1.12,
    showarrow=False,
    align="left",
    text=(
        "Position is the average Google rank for a page (1 is best). "
        "CTR is click-through rate: clicks / impressions. "
        "Pages toward lower position numbers with higher CTR are generally performing better."
    ),
)
fig_ctr.update_layout(margin={"t": 140})
lifeline_theme.add_lifeline_logo(fig_ctr)
fig_ctr.show()
```

## Optional: BigFrames quick check

If you prefer a pandas-like interface that still runs on BigQuery, this cell previews the same table with BigFrames.


```python
# Optional and safe: skip if bigframes is not installed in this environment.
try:
    import bigframes.pandas as bpd
except ModuleNotFoundError:
    print("bigframes is not installed; skipping BigFrames demo cell.")
else:
    bpd.options.bigquery.project = config.PROJECT_ID

    # Push date predicates into table read to align with partitioning on report_date.
    bf_df = bpd.read_gbq_table(
        f"{config.PROJECT_ID}.{config.SEARCHCONSOLE_DATASET}.seo_page_daily",
        index_col="report_date",
        filters=[
            ("report_date", ">=", str(window.start_date)),
            ("report_date", "<=", str(window.end_date)),
        ],
    )

    bf_top = (
        bf_df.groupby("page_path", as_index=False)[["gsc_clicks", "gsc_impressions"]]
        .sum()
        .sort_values("gsc_clicks", ascending=False)
        .head(10)
    )
    display(bf_top.to_pandas())
```
